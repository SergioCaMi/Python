```python
#Objeto iloc (Integer Location): localización por índice. Permite acceder a filas/Columnas de un DataFrame usando los índices numéricos.
#iloc[x, y] (fila y columna). Los índices empiezan por 0.

import pandas as pd


ruta = r"C:\Users\sergi\Desktop\Python\BBDD\VentasGroucery_Meses.xlsx"

ventas_agosto = pd.read_excel(ruta, sheet_name="Ventas Agosto 2014" ) 

ventas_septiembre = pd.read_excel(ruta, sheet_name="Ventas Septiembre 2014" ) 

ventas_octubre = pd.read_excel(ruta, sheet_name="Ventas Octubre 2014" ) 

#Creamos el trimestre
trimestre2014 = pd.concat([ventas_agosto, ventas_septiembre, ventas_octubre],ignore_index=True)
#Creamos la columna Amount
trimestre2014['Amount'] = (trimestre2014['Unit Price']* trimestre2014['Quantity'])

# Vamos a trabajar con iloc, que es de lectura y de escritura
#Cambiar el valor:
trimestre2014.iloc[1,1]= "Queso Cabrales"
trimestre2014

```

```python
# Con iloc puedo poner todas las filas y algunas de las columnas en el orden que yo quiera
trimestre2014.iloc[:, [5,3,6,1]]
```

```python

# Si Queremos saber filas y columnas y cuales:
trimestre2014.info()
# Queremos recorrer todo el DataFrame y para ello utilizaremos un bucle (for). Necesito saber el número de iteraciones, por lo tanto, nececsitamos saber el número de filas
print(len(trimestre2014)) # Con len() recibimos el numero de filas que tiene el DataFrame.
print("Columnas: ",len(trimestre2014.columns)) # Columnas


# se puede conocer el índice de la columna Category con:  
columnaCategory_indice = trimestre2014.columns.get_loc("Category")



#Vamos a recorrer el DataFrame e imprimir el producto y entre paréntesis, su categoría
filas = len(trimestre2014)
for i in range(0,filas):
   print(trimestre2014.iloc[i, 1] + " (" + trimestre2014.iloc[i,2]+")")
```

```python
#Como convertir y guardar en un fichero agrupaciones de un dataframe:
#Suma de importe por categoría

trimestre2014.groupby(['Category']).Amount.sum() #Agrupamos por categoría y sumamos el Amount
trimestre2014.groupby(['Category']).Amount.mean() #Agrupamos por categoría y hacemos el promedio de Amount
trimestre2014.groupby(['Category']).Amount.count() #Agrupamos por categoría y hacemos el número de pedidos
```

```python
#Como guardamos estos datos en un dataframe? 
#Primero creamos un objeto para guardar esta información
amount_category = trimestre2014.groupby(['Category']).Amount.sum()
amount_category

#Vamos a convertir este objeto en un DataFrame de pandas:
amount_category = pd.DataFrame(amount_category, columns=['Amount']) #Le decimos que Amount es otra columna. Solo en caso de error, no es necesario.
amount_category
#Al convertir el objeto en un DF, pandas cree que Category es su indice. Lo que debemos hacer es generar un nuevo indice numérico:
#reset_index:
amount_category = amount_category.reset_index()
amount_category
```

```python

#Conectarnos con el Excel:
ruta = r"C:\Users\sergi\Desktop\Python\BBDD\Practica Merge Pandas.xlsx"

#Cargar los dataframes:
facturas = pd.read_excel(ruta, sheet_name='Facturas')
clientes = pd.read_excel(ruta, sheet_name='Clientes')


facturas
clientes

#hacer el merge. Siempre se aplica a un objeto nuevo:

datos_completos = pd.merge(facturas, clientes, on="IdCliente", how="inner") #Inner Join
#qué dataframes unimos y por qué campo

#Solo muestra los registros que coinciden en ambos dataframes (en este caso, 4 y 5 no están en los dos) 
datos_completos

# esto es un INNER JOIN de SQL (Intersección, unicamente los registros coincidentes en los dos)
```

```python
#Si lo que queremos es ver TODAS las facturas aunque no nos coincida con clientes? allí donde no hay registro, saldrá en blanco NaN

datos_completos = pd.merge(facturas, clientes, on="IdCliente", how="left") #Left Join
datos_completos
```

```python
#Si lo que queremos es ver TODOS los clientes aunque no nos coincida con facturas? allí donde no hay registro, saldrá en blanco NaN

datos_completos = pd.merge(facturas, clientes, on="IdCliente", how="right") #Right Join
datos_completos
```

```python
#Si lo que queremos es ver la union completa: FULL OUTER JOIN. Mantendrá todos los registros de ambos DF.

datos_completos = pd.merge(facturas, clientes, on="IdCliente", how="outer") #FULL OUTER JO
datos_completos
```

```python
#Vamos a conectar con SQL Server mediante una cadena de conexión. Lo que hay que indicar es el servidor, BBDD y seguridad
#Necesitamos una librería para conectar con el servidor: py.odbc
#si no está, instalar.

import pyodbc 
import pandas as pd

server = "SERGIOPC\SQLEXPRESS"          #Nombre del servidor
database = "NEPTUNO"                    #Nombre de la base de datos
connection = 'yes'                      #Conexión
#La cadena de conexión
conn_str = (
    f'DRIVER={{SQL Server}};'
    f'SERVER={server};'
    f'DATABASE={database};'
    f'TRUSTED_CONNECTION={connection};'
)

#Control de errores y excepciones: sirve para interceptar un error y darle tratamiento.
try:
    #intenta este codigo:
    conn = pyodbc.connect(conn_str)
    print("*** Conectado con éxito! ***")
    #En SQL tenemos dos tipos de sentencias: 
    # la primera que modifica estructuras (CREATE (Crear BBDD, tablas, vistas, ...) y ALTER (Modificar) y DROP (Eliminar)). 
    # La segundas son las que trabajan con registros INSERT INTO, UPDATE, DELETE, SELECT 
    #SQL es casi igual en todos los casos.
    query = 'SELECT * FROM Categorías'
    print("OK", query)
    #Si la tabla tiene espacios en blanco en el nombre, va entre []
    query2 = 'SELECT * FROM [Detalles de pedidos]'
    print("OK", query2)

    #Vamos a cargar en un dataframe las categorías y en otro los detalles de los pedidos
    #Primer parámetro: la sentencia SQL; segundo parámetro: el objeto que tiene la BBDD; 
    categorias = pd.read_sql(query, conn)
    display(categorias)
    detalles = pd.read_sql(query2, conn)
    display(detalles)
except Exception as e:
    #En caso de error, lo guarda en la variable e y hace las siguientes acciones:
    print("Fallo de conexión! ", e)



```
